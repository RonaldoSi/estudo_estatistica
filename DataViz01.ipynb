```python
# criar gráficos com quatro subplots

fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.suptitle('Taxas de imigração dos quatro maiores países da América do Sul para o Canadá')
fig.subplots_adjust(hspace=0.5, wspace=0.3)
# hspace -> ajustar espaço horizontal
# wspace -> ajustar espaço vertical
# assim dando mais espaçamento entre os gráficos

axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')


# realizando o ajuste no xlabel de forma conjunta com "for", onde os valores estavam todos sobrepostos
for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))

# setando títulos nos eixos de forma cojunta em todos os gráficos
for ax in axs.flat:
  ax.set_xlabel('Ano')
  ax.set_ylabel('Número de imigrantes')

ymin = 0
ymax = 7000

for axs in axs.ravel():
  axs.set_ylim(ymin, ymax)

plt.show()
```

# **Exercício 02**

Você trabalha como Analista de Dados em uma empresa de varejo e recebeu a tarefa de criar uma figura com subplots que apresente a variação no número de vendas em quatro diferentes lojas ao longo de um ano. A gerência da empresa precisa visualizar de forma clara as tendências de vendas em cada loja, para que possam tomar decisões estratégicas sobre os estoques e ações de marketing. Para isso, você deve criar quatro subplots dispostos em duas linhas e duas colunas, onde cada subplot representa uma loja diferente. Nesse desafio, cada subplot deve apresentar um gráfico de linhas que mostre a variação do número de vendas ao longo dos meses do ano.

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
df_desafio = pd.DataFrame(vendas_2022, index=lojas)
df_desafio
```

```python
fig, axs = plt.subplots(2,2, figsize=(12,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Indíce de vendas para as lojas A, B, C e D')

axs[0,0].plot(df_desafio.loc['A'])
axs[0,0].set_title('Loja A')

axs[0,1].plot(df_desafio.loc['B'])
axs[0,1].set_title('Loja B')

axs[1,0].plot(df_desafio.loc['C'])
axs[1,0].set_title('Loja C')

axs[1,1].plot(df_desafio.loc['D'])
axs[1,1].set_title('Loja D')

for ax in axs.flat:
  ax.set_xlabel('Mês')
  ax.set_ylabel('Quantidade de vendas')
```

# **Retomando o conteúdo...**

Aprendendo alguns parâmetros, como:

lw -> aumenta a espessura das linhas do gráfico

marker -> realiza a plotagem de marcadores no gráfico

loc (título) -> usando para definirmos a disposição do título

grid -> adiciona linhas de grade ao gráfico

linestyle -> podemos passar parâmetros para modificar as linhas de grade, exemplo("--")

```python
# criando um gráfico como figura
fig, ax = plt.subplots(figsize=(8,4))

ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=3, marker='o', color='green') # lw -> aumentar espessura da linha
ax.set_title('Imigração do Brasil para o Canadá\n1980-2013', fontsize=18, loc='left') # loc -> serve para alterarmos onde o título vai ficar disposto
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Número de imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12) # alterando o tamanho dos valores do eixo x
ax.yaxis.set_tick_params(labelsize=12) # alterando o tamanho dos valores do eixo y
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # colocando os anos de 5 em 5 de forma mais fácil
plt.grid(linestyle='--') # adiciona grade ao gráfico

plt.show()
```

Adicionando cores em várias variáveis de um gráfico

```python
df.head()
```

```python
paises_america_do_sul = df.query('Região == "América do Sul"')
paises_america_do_sul
```

```python
america_sul_sorted = paises_america_do_sul.sort_values('Total', ascending=True) # ordenando os valore de total para aperecer de forma ordenada no gráfico
```

```python
cores = []

for pais in america_sul_sorted.index:
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('silver')

fig, ax = plt.subplots(figsize=(12,5))

ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('Imigração da américa do Sul para o Canadá\n1980-2013', loc='left', fontsize=18)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

# adicionando rótulo de dados
for i, v in enumerate(america_sul_sorted['Total']):
  ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')

# retirando o contorno do gráfico
ax.set_frame_on(False)

# tirando os valores do eixo x
ax.get_xaxis().set_visible(False)

# tirando uma pequena marcação nas barras no eixo y
ax.tick_params(axis='both', which='both', length=0)

fig.savefig('migracao_america_sul_canada', transparent=False, dpi=300, bbox_inches='tight')

plt.show()
# print(fig.canvas.get_supported_filetypes()), comando para verificar os tipos de arquivos disponíveis para salvar a imagem
```

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```
